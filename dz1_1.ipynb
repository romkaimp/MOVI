```python
h = 0.2
a, b, c = 1, 4, 2.5

def a2(a, b, c, f: Callable):
    return 1/(b-c) * ( (f(b) - f(a))/(b - a) - (f(c) - f(a))/(c - a))

def a1(a, b, c, f: Callable):
    return (f(c) - f(a)) / (c - a)

i =0
while True:
    print("i:", i)
    print(f"a, c, b: {a:.5f}, {c:.5f}, {b:.5f}")
    print(f"f(a), f(c), f(b): {f(a):.5f}, {f(c):.5f}, {f(b):.5f}")
    i += 1
    if (c - a)/(b - a) < h:
        x_min = c + (b - c) / 4
        break
    elif (b - c)/(b - a) < h:
        x_min = c - (c - a) / 4
        break
    x_min = 1/2 * (a + c - a1(a, b, c, f) / a2(a, b, c, f))

    if x_min < c:
        if f(x_min) > f(c):
            a, c, b = x_min, c, b
        else:
            a, c, b = a, x_min, c
    else:
        if f(x_min) > f(c):
            a, c, b = a, c, x_min
        else:
            a, c, b = c, x_min, b

print(f"x_min: {x_min:.5f}, f(x_min): {f(x_min):.5f}")

print(f"{f(a):.5f}, {a1(a, b, c, f):.5f}, {a2(a, b, c, f):.5f}")
```
